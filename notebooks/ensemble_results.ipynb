# Ensemble results

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

## Curva de entrenamiento

```python
exp = 'flatten_linear_ensemble_hNone_lr0.01_ep500_bias_test_01012026-214853'
dataset = 'mini'
train_log_path = Path('../models') / dataset / f'{exp}_training_log.csv'
train_log = pd.read_csv(train_log_path)
```

```python
train_log.head()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
epoch_lim = 100

# Loss
ax1.plot(train_log['epoch'], train_log['loss'], c='#f46524')
ax1.set_xlabel('poca')
ax1.set_xlim(0, epoch_lim)
ax1.set_ylabel('Loss')
ax1.set_title('Loss durante el entrenamiento (con dev)')

# Plot accuracy
ax2.plot(train_log['epoch'], train_log['accuracy'], c='#f46524')
ax2.set_xlabel('poca')
ax2.set_xlim(0, epoch_lim)
ax2.set_ylabel('Accuracy')
ax2.set_title('Accuracy durante el entrenamiento (con dev)')

plt.tight_layout()
plt.show()
```
